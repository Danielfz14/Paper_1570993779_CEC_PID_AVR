```python
class P1(bf.BasicProblem):
    def __init__(self, variable_num):
        super().__init__(variable_num)
        self.max_search_range =np.array([1.5, 1.5, 1,1.5, 1.5, 1 ])                    
        self.min_search_range = np.array([0.000001, 0.0000001,0.0000001,0.000001, 0.0000001,0.0000001]) 
        self.global_optimum_solution = 0.0
    def get_func_val(self, variables, *args):
        fcost=eng.AVR_PID_PID_CEC_p1(float(variables[0]),float(variables[1]),float(variables[2]),float(variables[3]),float(variables[4]),float(variables[5]))
        return fcost 
```
